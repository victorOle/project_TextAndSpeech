# Spacy model approach

```python
import re
import xml.etree.ElementTree as ET
import spacy

# Load spaCy model for chemical entity recognition
nlp = spacy.load("en_core_sci_md")

# Define a function for text preprocessing
def preprocess_text(text):
    text = re.sub(r'\[(\d*[,\-\s])*\d*\]', '', text)  # remove references
    #text = re.sub(r'\W', ' ', text)  # Remove non-alphanumeric characters
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespaces
    text = text.lower()  # Convert to lowercase
    return text

# Function to extract chemical entities from text using spaCy
def extract_chemical_entities(text):
    doc = nlp(text)
    chemical_entities = [ent.text for ent in doc.ents]# if ent.label_ == "CHEMICAL"]
    return chemical_entities

def extract_annotation(passage):
    annotation_types = []
    identifiers = []
    entities = []
    offsets = []
    for annotation in passage.findall('annotation'):
        annotation_types.append(annotation.find('infon[@key="type"]').text)
        identifiers.append(annotation.find('infon[@key="identifier"]').text)
        entities.append(annotation.find('text').text)
        offsets.append(annotation.find('location').get("offset"))
    return (annotation_types, identifiers, entities, offsets)

# Function to process XML content and perform chemical entity recognition
def process_xml(xml_file_path):
    #tree = BeautifulSoup(xml_content, 'html.parser')

    tree = ET.parse(xml_file_path)
    doc = tree.getroot().find('document')
    
    for passage in doc.findall('passage'):
        passage_type = passage.find('infon[@key="type"]').text
        # skip titles
        if "title" in passage_type:
            continue
        text = passage.find('text').text
        
        preprocessed_text = preprocess_text(text)
        chemical_entities = extract_chemical_entities(preprocessed_text)
        annotation_types, identifiers, annotation_entities, offsets = extract_annotation(passage)

        # Print results
        print(f"Passage Type: {passage_type}")
        print(f"Original Text: {text}")
        print(f"Preprocessed Text: {preprocessed_text}")
        print(f"Chemical Entities: {chemical_entities}")
        print(f"Annotated Entities: {annotation_entities}")
        print("\n")

# Sample XML content (replace with your actual XML content)
xml_file_path = "data/ALL/555756_v1.xml"

# Process the XML content
process_xml(xml_file_path)
```
